```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df['Attrition'].value_counts(normalize=True)
```

```python
sns.set_style('whitegrid')
```

```python
attrition = df['Attrition'].value_counts(normalize=True)
plt.figure(figsize=(8, 6))
sns.barplot(x=attrition.index, y=attrition)
plt.title('Distribution Of Attrition Rate')
plt.xlabel('Attriton')
plt.ylabel('Percentage')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
attrition = sns.barplot(data=df, x='YearsAtCompany', y='Age', hue='Attrition',dodge=True, palette='coolwarm')

# Set labels and title
plt.xlabel("Years at Company")
plt.ylabel("Average Age")
plt.title("YearsAtCompany vs Attrition (Age as Color)")
plt.legend(title="Attrition")

# Show the plot
plt.show()
```

```python
avg_tenure = df['YearsAtCompany'].mean()
print(f'Average years of employee to leave the company is {avg_tenure} years')
 
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
sns.histplot(data=df, x='Age', kde=True, ax=axes[0])
axes[0].set_title('Distribution Employee by Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')
sns.countplot(data=df, x='Gender', ax=axes[1])
axes[1].set_title('Distribution Employee by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Count')
sns.countplot(data=df, x='Department', ax=axes[2])
axes[2].set_title('Distribution Employee by Department')
axes[2].set_xlabel('Department')
axes[2].set_ylabel('Count')
plt.tight_layout()
plt.show()
```

TotalWorkingYears vs. DailyRate (with Education as Category)

```python
# Scatter plot with trend line
sns.lmplot(data=df, x='TotalWorkingYears', y='DailyRate', hue='Education', aspect=1.5)
plt.title("TotalWorkingYears vs. DailyRate (Education as Category)")
plt.show()
```

Years Since Last Promotion vs Attrition (WorkLifeBalance as Color)

```python
# Create heatmap
pivot_table = df.pivot_table(index='YearsSinceLastPromotion', columns='Attrition', values='WorkLifeBalance', aggfunc='mean')
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Years Since Last Promotion vs Attrition (WorkLifeBalance as Color)")
plt.xlabel("Attrition")
plt.ylabel("Years Since Last Promotion")
plt.show()
```

Attrition vs Education Field

```python
# Create bar chart
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='EducationField', hue='Attrition', palette='coolwarm')
plt.title("Attrition vs EducationField")
plt.xlabel("Education Field")
plt.ylabel("Count")
plt.legend(title="Attrition")
plt.show()
```

Daily Rate across Business Travel Categories

```python
# Create box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='BusinessTravel', y='DailyRate', palette='coolwarm')
plt.title("DailyRate across BusinessTravel Categories")
plt.xlabel("Business Travel")
plt.ylabel("Daily Rate")
plt.show()
```

Years At Company vs Total Working Years

```python
# Create line chart
plt.figure(figsize=(8, 6))
sns.lineplot(data=df, x='YearsAtCompany', y='TotalWorkingYears', marker='o', color='blue')
plt.title("YearsAtCompany vs TotalWorkingYears")
plt.xlabel("Years at Company")
plt.ylabel("Total Working Years")
plt.show()
```
